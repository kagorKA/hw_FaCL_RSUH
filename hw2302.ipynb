Задача 3.

Представим себе, что мы планируем написать игру "Тамагочи". Напишите класс Cat для этой игры.

Imagine that we're planning to write a game Tamagochi. Define a Cat class for this game.

```python
import time

class Cat:

    def __init__(self, name, age=0, relation=10, health=50, stamina=50, howmchslp=0):
        self.name = name
        self.age = age
        self.relation = relation
        self.health = health
        self.stamina = stamina
        self.howmchslp = howmchslp
    def __str__(self):  
        if self.age == 1:
          return f'This is {self.name} it is {self.age} year old\nRelationship: {self.relation}%\nHealth: {self.health}\nStamina: {self.stamina}'   
        else:
          return f'This is {self.name} it is {self.age} years old\nRelationship: {self.relation}%\nHealth: {self.health}\nStamina: {self.stamina}'
    def __repr__(self):
        return f"Cat: '{self.name}', {self.age}, {self.relation}, {self.health}, {self.stamina}"        
    
    def speak(self):
        print('Meow')

    def feed(self):
        if self.health == 100:
          print(f'{self.name} is full')
        else:
          self.health += 50
          self.stamina -= 10
          print(f'Good recharge! {self.name}\'s health is {self.health}')
          if self.stamina == 0:
            print(f'Now it\'s time for {self.name} to go to sleep')
    
    def sleep(self):
        if self.stamina > 0:
          print(f'It\'s too early to sleep')
        else:
          time.sleep(10)
          self.stamina += 100
          print(f'Good! {self.name} is ready to play!')
          self.howmchslp += 1
          self.grow()
    
    def play(self):
        if self.health == 0:
          print(f'{self.name} is too hungry to play now :(')
        elif self.stamina == 0:
          print(f'{self.name} is too tired to play now :(')
        elif self.health == 0 and self.stamina == 0:
          print(f'{self.name} is too tired and hungry to play now. Help your cat!')
        else:
          if self.relation < 100:
            self.relation += 10
            self.health -= 50
            self.stamina -= 10
            print(f'{self.name} is happy!\nYour relationship is {self.relation}% now.\nIt\'s time to feed your cat')
          else:
            self.health -= 50
            self.stamina -= 10
            print(f'{self.name} loves you for {self.relation}%, it is happy!\nNow feed your cat')

    def grow(self):
      if self.howmchslp % 3 == 0 and self.age < 18:
        self.age += 1
        print(f'Your cat has grown up. It has {self.age} years now')
      elif self.howmchslp % 3 == 0 and self.age == 18:
        print(f'{self.name} RIP X__X')
```
